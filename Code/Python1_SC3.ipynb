```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import Hillsborough Data
df = pd.read_excel('Combined.xlsx', sheet_name = 'Hillsborough')
```

```python
#Dataframe and Graph Moving Average and Daily Cases and Set Axes
df.columns = ['Date', 'Daily_Number', 'Moving_Average']
axes = df.plot('Date','Daily_Number', color='red',xticks=[])
df.plot('Date', 'Moving_Average', kind='bar',xticks=[], width=1, ax=axes)
```
